# Closed Restaurant Exploration

In another notebook we assigned businesses to one of the ten metro areas and screened out businesses that didn't seem to belong to any of those metros.  This now sits in a nice JSON file bus_metro2.json.

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
import json
import numpy as np
import pandas as pd
```

```python
with open("bus_metro2.json", "rt") as infile:
    dfb = pd.read_json(infile)
```

```python
dfb
```

Create a dataframe of closed restaurants in Phoenix. Create a second dataframe of open businesses in Phoenix

```python
closed_phx = dfb[(dfb.open == False) & (dfb.categories.map(lambda x: "Restaurants" in x))& (dfb.metro == 'Phoenix')]
print(len(closed_phx))
open_phx = dfb[(dfb.open == True) & (dfb.metro == 'Phoenix')]
print(len(open_phx))
```

Now we have 2610 closed restaurants and 31186 open businesses. How many open businesses are located at these closed restaurants?

Merge the two dataframes based on the full address with an inner merge, only keeping what shows up in both dataframes

```python
reopen_phx = pd.merge(closed_phx, open_phx, how = 'inner', left_on = 'full_address', right_on = 'full_address')
print(len(reopen_phx))
print(len(reopen_phx.full_address.unique()))
```

```python
reopen_phx.sort_values('name_x')
```

Now there are 2218 businesses that have reopened at addresses of closed restaurants. However, many are duplicate. We see there are 819 businesses at unique addresses, so there should be 2610-819 = 1791 businesses where no business has opened at that address. This also leaves open the possibility that the same building now has a different address.

But really we want the number of unique combinations of address and name.

```python
print(len(reopen_phx.groupby(['full_address','name_x'])))

print(len(closed_phx.groupby(['full_address','name'])))
```

So there are 2,606 unique combinations of address and business name that are closed in the database. 
There are 1,014 unique combinations of address and business name of closed businesses that match addresses of open businesses.

So, 2606 - 1014 = 1592 is the anticipated number of closed businesses that should still be closed.

How do we get the list of closed restaurants that have stayed closed? Let's try a left merge of the original closed restaurants with the reopened list and filter out anything that doesn't join. 

```python
unopen_phx = pd.merge(closed_phx, reopen_phx, how = 'left', left_on = 'full_address', right_on = 'full_address')
print(len(unopen_phx[unopen_phx.metro_y.isnull()].groupby(['full_address','name'])))
```

Now we have the anticipated 1,592 records. Let's look at them
